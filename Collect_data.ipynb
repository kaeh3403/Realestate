```python
import requests, bs4
import pandas as pd
from lxml import html
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus, unquote

xmlUrl = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev'

API_Key = unquote("SzUttTWCXo4WbRF7PiAE4y3rZ5Lnhwc5diiuQxqy6EHyZO4GfWN%2FHGSCveUjD7uMjh%2FD6rGAx977eKuZUY376A%3D%3D")
```

```python
def collect_data(serviceKey, sigunguCode, yearMonth):

    xmlUrl = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcNrgTrade'

    API_Key = unquote(serviceKey)


    
    queryParams = "?" + urlencode(
        {
            quote_plus("ServiceKey") : API_Key,
            quote_plus("LAWD_CD") : sigunguCode,
            quote_plus("DEAL_YMD") : yearMonth
        }
    )

    response = requests.get(xmlUrl + queryParams).text.encode('utf-8')
    xmlobj = bs4.BeautifulSoup(response, 'lxml-xml')
    rows = xmlobj.findAll("item")
    if len(rows)==0:
        print("결과가 조회되지 않음.")
        print(sigunguCode, yearMonth)
        return [sigunguCode, yearMonth]
    columns = rows[0].find_all()

    rowList = []
    nameList = []
    columnList = []

    rowsLen = len(rows)
    for i in range(0, rowsLen):
        columns = rows[i].find_all()

        columnsLen = len(columns)
        for j in range(0, columnsLen):
            if i == 0:
                nameList.append(columns[j].name)
            eachColumn = columns[j].text
            columnList.append(eachColumn)
        rowList.append(columnList)
        columnList = []

    result = pd.DataFrame(rowList, columns= nameList)
    return result
```
