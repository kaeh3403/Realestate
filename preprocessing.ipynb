```python
import os
import pandas as pd
```

```python

path = "./인천/분양입주권/"
file_list = os.listdir(path)
```

```python
data = pd.DataFrame()
for i in file_list:
    file_name = path+i
    df = pd.read_csv(file_name)
    data = pd.concat([data, df])
```
